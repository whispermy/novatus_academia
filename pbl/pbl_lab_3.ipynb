```python
pos_directory = '/content/drive/MyDrive/review_polarity/txt_sentoken/pos/'
neg_directory = '/content/drive/MyDrive/review_polarity/txt_sentoken/neg/'
```

```python
from os import listdir
```

```python
pos_list = listdir(pos_directory)
neg_list = listdir(neg_directory)
```

```python
pos_data = {}
neg_data = {}

for name in pos_list:
  file_temp = open(pos_directory+name, 'r')
  text_temp = file_temp.read()
  file_temp.close()
  token_temp = text_temp.split()
  token_lst_temp = []
  for i in token_temp:
    temp = i.translate(table)
    if len(temp) > 1 and temp not in stopwords and temp.isalpha():
      token_lst_temp.append(temp)
    
  pos_data[name] = token_lst_temp

for name in neg_list:
  file_temp = open(neg_directory+name, 'r')
  text_temp = file_temp.read()
  file_temp.close()
  token_temp = text_temp.split()
  token_lst_temp = []
  for i in token_temp:
    temp = i.translate(table)
    if len(temp) > 1 and temp not in stopwords and temp.isalpha():
      token_lst_temp.append(temp)
    
  neg_data[name] = token_lst_temp
```

```python
from collections import Counter
```

```python
vocab = Counter()
```

```python
# 모든 값에 대해서 counter에 update하기

for i in pos_data.keys():
  vocab.update(pos_data.get(i))

for i in neg_data.keys():
  vocab.update(neg_data.get(i))


print(vocab)
```

```python
# 상위 50개 데이터 추출하기
sort_vocab = sorted(vocab.items(), key = lambda x : x[1], reverse=True)
vocab_50 = sort_vocab[:50]


# 횟수가 2 이상인 데이터만 단어 추출하기
# words = []
# for key, values in vocab.items():
#   if values > 1:
#     words.append(key)

myvocab = [ k for k,v in vocab.items() if v > 1]
print(len(myvocab))
```

```python
# 텍스트를 1줄씩 나누고, 학습을 위한 x 데이터 생성하기

sentence = []
s = 0
for name in pos_list:
  file_temp = open(pos_directory+name, 'r')
  text_temp = file_temp.read()
  file_temp.close()
  lines = text_temp.splitlines()
  s += len(lines)
  for line in lines:
    new_line = line.split()
    rst = [ i.translate(table) for i in new_line if i.translate(table) in myvocab]
    if rst != []:
      sentence.append(rst)    
      # print(rst)
print("pos_line sum:",s)
s = 0
for name in neg_list:
  file_temp = open(neg_directory+name, 'r')
  text_temp = file_temp.read()
  file_temp.close()
  lines = text_temp.splitlines()
  s += len(lines)
  for line in lines:
    new_line = line.split()
    rst = [ i.translate(table) for i in new_line if i.translate(table) in myvocab]
    if rst != []:
      sentence.append(rst)    
      # print(rst)
print("pos_line sum:",s)
```
