```python
import tensorflow.keras as keras
from os import listdir
import numpy as np
```

```python
file_path = '/content/drive/MyDrive/txt_sentoken/vocab.txt'
file_1 = open(file_path, 'r')
vocab_text = file_1.read()
file_1.close()

file_path = '/content/drive/MyDrive/txt_sentoken/embedding_word2vec.txt'
file_1 = open(file_path, 'r')
w2v_text = file_1.readlines()[1:]
file_1.close()
```

```python
print(w2v_text[:10])
```

```python
w2v_dict = dict()
for line in w2v_text:
  part = line.split()
  w2v_dict[part[0]] = np.asanyarray(part[1:], dtype='float32')
```

```python
print(w2v_dict['film'])
```

```python
vocab_key = w2v_dict.keys()
vocab_key
```

```python
import string
from nltk.corpus import stopwords
import nltk

nltk.download('stopwords')
stopwords = set(stopwords.words('english'))

table = str.maketrans(" "," ",string.punctuation)

pos_directory = '/content/drive/MyDrive/review_polarity/txt_sentoken/pos/'
neg_directory = '/content/drive/MyDrive/review_polarity/txt_sentoken/neg/'

pos_list = listdir(pos_directory)
neg_list = listdir(neg_directory)

pos_data = {}
neg_data = {}

for name in pos_list:
  file_temp = open(pos_directory+name, 'r')
  text_temp = file_temp.read()
  file_temp.close()
  token_temp = text_temp.split()
  token_lst_temp = []
  for i in token_temp:
    temp = i.translate(table)
    if len(temp) > 1 and temp not in stopwords and temp.isalpha():
      token_lst_temp.append(temp)
    
  pos_data[name] = token_lst_temp

for name in neg_list:
  file_temp = open(neg_directory+name, 'r')
  text_temp = file_temp.read()
  file_temp.close()
  token_temp = text_temp.split()
  token_lst_temp = []
  for i in token_temp:
    temp = i.translate(table)
    if len(temp) > 1 and temp not in stopwords and temp.isalpha():
      token_lst_temp.append(temp)
    
  neg_data[name] = token_lst_temp
```

```python
pos_train_data = []
neg_train_data = []

for i in pos_data.keys():
  L = []
  for k in pos_data[i]: 
    if k in vocab_key:
      L.append(k)
  cleaned_text = ' '.join(L)
  pos_train_data.append(cleaned_text)

for i in neg_data.keys():
  L = []
  for k in neg_data[i]: 
    if k in vocab_key:
      L.append(k)
  cleaned_text = ' '.join(L)
  neg_train_data.append(cleaned_text)
```

```python
def make_train_data(posneg):
  tr_data = []
  directory = '/content/drive/MyDrive/review_polarity/txt_sentoken/' + posneg
  lst = listdir(directory)
  data = {}

  for name in lst:
    file_temp = open(directory+name, 'r')
    text_temp = file_temp.read()
    file_temp.close()
    token_temp = text_temp.split()
    token_lst_temp = []

    for i in token_temp:
      temp = i.translate(table)
      if len(temp) > 1 and temp not in stopwords and temp.isalpha() and temp in vocab_key:
        token_lst_temp.append(temp)
    
    cltext = ' '.join(token_lst_temp)
    tr_data.append(cltext)

  return tr_data
```

```python
func_pos_data = make_train_data('pos/')
func_neg_data = make_train_data('neg/')
```

```python
func_train_data = func_pos_data + func_neg_data
```

```python
func_train_data[0]
```

```python
from keras.preprocessing.text import Tokenizer
from keras.preprocessing.sequence import pad_sequences
```

```python
tokenizer = Tokenizer()
tokenizer.fit_on_texts(func_train_data)
```

```python
print(tokenizer.word_index)
```

```python
embedding_vector = np.zeros((len(tokenizer.word_index)+1,100))
```

```python
for name,num in tokenizer.word_index.items():
  embedding_vector[num] = w2v_dict[name]
```

```python
print(embedding_vector[len(tokenizer.word_index)])
```

```python
encoded_docs = tokenizer.texts_to_sequences(func_train_data)
```

```python
print(encoded_docs[0])
print(func_train_data[0])
```

```python
max_length = len(max(func_train_data))
max_length
```

```python
line = 0
new_line = 0
for i in func_train_data:
  line = len(i)
  if line > new_line:
    new_line = line

L = []
for i in func_train_data:
  L.append(len(i))

print(new_line, max(L))
max_l = max(L)
```

```python
Xtrain = keras.preprocessing.sequence.pad_sequences(encoded_docs, maxlen=max_l, padding='post')
Ytrain = np.array([0 for _ in range(1000)] + [1 for _ in range(1000)])
```

```python
import tensorflow.keras as keras

size = len(tokenizer.word_index)+1

embedding_layer = keras.layers.Embedding(size, 100, weights=[embedding_vector], input_length=max_l, trainable=False)

model = keras.models.Sequential()
model.add(embedding_layer)
model.add(keras.layers.Conv1D(filters=128, kernel_size=5, activation='relu'))
model.add(keras.layers.MaxPooling1D(pool_size=2))
model.add(keras.layers.Flatten())
model.add(keras.layers.Dense(1, activation='sigmoid'))

print(model.summary())
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
model.fit(Xtrain, Ytrain, epochs=10)
```
