```python
#input 한글자씩 분리시키기 
list_input = []

for i in range(len(data['INPUT'])):
  sp = data['INPUT'][i]
  new_line = ''
  for j in sp:
    # if j.isalpha():
    #   j = j.upper()
    if j != ' ':
      new_line += ' '
      new_line += j
  list_input.append(new_line[1:])
```

```python
#또는 밑에서 배운 형태소 분석기로 분리시키기
list_input_kor = []

for i in range(len(data['INPUT'])):
  txt = data['INPUT'][i]
  new_line = []
  t = okt.pos(txt,norm=True, stem=False, join=False)
  for k in t:
    new_line.append(k[0])
  list_input_kor.append(new_line)
```

```python
#가중치 행렬
import numpy as np

def get_weight_matrix(embedding, vocab):

	# total vocabulary size plus 0 for unknown words
  vocab_size = len(vocab) + 1
  weight_matrix = np.zeros((vocab_size, 2))
	# step vocab, store vectors using the Tokenizer's integer mapping
  for word, i in vocab.items():
    if word.isalpha():
      word = word.upper()
    weight_matrix[i] = embedding.get(word)
  return weight_matrix

embedding_vectors = get_weight_matrix(input_vector_dic, tokenizer.word_index)
embedding_vectors.shape
embedding_vectors[np.isnan(embedding_vectors)] = 0
```
